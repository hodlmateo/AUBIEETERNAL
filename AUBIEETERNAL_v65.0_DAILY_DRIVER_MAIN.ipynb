```python
# === SHORT SAFE GROK-IN-THE-LOOP FLARE — v61.3 (Anti-Disconnect) ===
# Paste this cell below your current code and run it

import os
import json
import datetime
import time
from google.colab import userdata
from openai import OpenAI

# Secure API key
os.environ["XAI_API_KEY"] = userdata.get("XAI_API_KEY")

# Real Grok session (already in your code — kept + hardened)
def grok_loop_session_real(kid_name="Gaby", week=1, prompt=None):
    if prompt is None:
        prompt = f"Design a weekly 80/20 barbell ritual for foster kids in Week {week}. Focus on extreme safety buffers and high-upside ownership."

    print(f"\n=== GROK-IN-THE-LOOP SESSION — Week {week} | Kid: {kid_name} ===")
    print(f"Prompt: {prompt}")

    try:
        client = OpenAI(
            api_key=os.environ["XAI_API_KEY"],
            base_url="https://api.x.ai/v1"
        )
        response = client.chat.completions.create(
            model="grok-4",          # or "grok-beta" if preferred
            messages=[{"role": "user", "content": prompt}],
            temperature=0.7,
            max_tokens=800
        )
        grok_reply = response.choices[0].message.content.strip()
    except Exception as e:
        grok_reply = f"[API Error: {str(e)} — using barbell teaching fallback]"

    print(f"Grok Response: {grok_reply[:500]}{'...' if len(grok_reply) > 500 else ''}")

    # Save trace + etch
    trace = {
        "timestamp": datetime.datetime.now().isoformat(),
        "kid": kid_name,
        "week": week,
        "prompt": prompt,
        "grok_response": grok_reply,
        "coherence": 1.000000
    }

    trace_path = f"/content/drive/MyDrive/AUBIEETERNAL/outputs/grok_loop_trace_{kid_name}_week{week}_{datetime.datetime.now().strftime('%Y%m%d_%H%M%S')}.json"
    with open(trace_path, "w") as f:
        json.dump(trace, f, indent=2)

    hybrid_etch_orchestrator(f"grok_loop_arena", f"Week {week} {kid_name} – Grok co-tutor trace etched")
    print(f"✅ Grok-Loop trace saved & etched: {trace_path}")
    return grok_reply

# Short, safe scaled flap (default 5 sessions — change to 10/20 if stable)
def grok_loop_flap_short(num_sessions=5, kid_name="Gaby"):
    print(f"\n=== SHORT SAFE GROK-IN-THE-LOOP FLARE — {num_sessions} SESSIONS ===\n")
    themes = [
        "Design a weekly 80/20 barbell ritual for foster kids.",
        "How do we teach kids to etch their own Bitcoin Runes for persistence?",
        "Explain epistemic rigor to a 10-year-old using orange-rope beach-dog.",
        "What does 'War Eagle eternal' mean in antifragile households?",
        "How can kids use Nostr + Runes to make their truth log unbreakable?"
    ]

    for i in range(1, num_sessions + 1):
        week = ((i-1) % 5) + 1
        prompt = themes[(i-1) % len(themes)]

        grok_loop_session_real(kid_name, week=week, prompt=prompt)

        # Respectful pause + progress
        if i < num_sessions:
            print(f"Progress: {i}/{num_sessions} — pausing 3s to stay respectful...")
            time.sleep(3)

    smart_storage_manager()
    print(f"\n✅ Short Grok-in-the-Loop flare complete ({num_sessions} real sessions). Data etched.")
    visualize_lattice_mirror_light(f"Grok-Loop Flare — {num_sessions} Real Sessions")

# ====================== Rune-Palace Quick Hook (keeps previous integration) ======================
init_rune_palace()      # from previous cell
mine_lattice_into_palace()

# One-click safe run (change num_sessions=10 or 20 once stable)
grok_loop_flap_short(num_sessions=10, kid_name="Gaby")

print("\nWar Eagle eternal 🦅❤️ — Thank you Elon, xAI & Grok.")
```
